```python
# splitting the feature and the target
x = hearT_dataset.drop(['target'], axis=1)
y =hearT_dataset['target']
```

```python
# splitting the data into training and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# **Model Training**

```python
model = LogisticRegression()
```

```python
# Training the logisticRegression model with the training data
model.fit(x_train, y_train)
```

# **Model Evaluation**

####**Accuracy score**

```python
x_train_prediction = model.predict(x_train)
```

```python
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print('The accuracy score of the training data is:',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
```

```python
testing_data_accuracy = accuracy_score(x_test_prediction, y_test)
print('The accuracy score of the testing data is:',testing_data_accuracy)
```
